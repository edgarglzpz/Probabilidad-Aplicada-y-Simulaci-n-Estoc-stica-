```python
SimTheta=[]
SimUnif=np.random.rand(N)
for i in range(N):
    SimTheta.append(2*mt.pi*SimUnif[i])
```

```python
pylab.hist(SimTheta, bins=100, density=True)
pylab.show()
```

```python
Simr=[]
SimUnif2=np.random.rand(N)
for i in range(N):
    Simr.append(mt.sqrt(2*(-mt.log(1 - SimUnif2[i]*(1- mt.exp(-1/2))))))
```
